```python
import numpy as np
```

```python
v1 = np.array([2, 3])
v2 = np.array([4, 1])
res = v1 + v2
print(res)
```

```python
# matrix multiplication and vector dot product
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
res = np.dot(A, B)
print(res)
```

```python
# transpose
transpose_A = A.T
print(transpose_A)
```

```python
# inverse
inv_A = np.linalg.inv(A)
print(inv_A)
```

### Rotation

```python
theta = np.radians(90)
R = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])
v = np.array([4, 2])
rotated_vector = np.dot(R, v)
print(rotated_vector)
```

```python
import matplotlib.pyplot as plt

def plot_vec(vectors, color, title):
  plt.axhline(0, color = 'black', linewidth = 0.5)
  plt.axvline(0, color = 'black', linewidth = 0.5)
  plt.grid()

  for i, vector in enumerate(vectors):
    plt.quiver(0, 0, vector[0], vector[1], angles = 'xy', scale_units = 'xy', scale = 1, color = color[i])
  plt.xlim(-10, 10)
  plt.ylim(-10, 10)
  plt.title(title)
  plt.show()

plot_vec([v, rotated_vector], ['blue', 'red'], ['Rotated by 90'])

#original is blue rotated is red
```

### Scaling

```python
# scale by 2x and 3y

v = np.array([4, 2])
s = np.array([2, 3])
res = v * s

#or s = np.array([[2, 0], [0, 3]])

plot_vec([v, res], ['blue', 'red'], 'Scaling (2x, 3y)')
```

### Translation

```python
def plot_translation(vectors, translation):
  plt.axhline(0, color = 'black', linewidth = 0.5)
  plt.axvline(0, color = 'black', linewidth = 0.5)
  plt.grid()

  new_vector = vectors + translation
  plt.quiver(0, 0, vectors[0], vectors[1], angles = 'xy', scale_units = 'xy', scale = 1, color = 'blue', label = 'original')
  plt.quiver(0, 0, new_vector[0], new_vector[1], angles = 'xy', scale_units = 'xy', scale = 1, color = 'red', label = 'translated')
  plt.xlim(-10, 10)
  plt.ylim(-10, 10)
  plt.legend()
  plt.title('Translation shift by 5, 4')
  plt.show()
```

```python
v = np.array([4, 2])
translation = np.array([5, 4])
plot_translation(v, translation)
```
